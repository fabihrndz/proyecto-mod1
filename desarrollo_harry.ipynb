```python
import random
```

```python
# Para contar los puntos
contador1 = 0
contador2 = 0
```

```python
# Creamos los hechizos
a=[
    "expelliarmus",
    "petrificus totalus",
    "imperio",
    "reducto",
    "crucio",
    "avada Kedavra",
    "sectumsempra",]
```

```python
d=[
    "repello",
    "expecto patronum",
    "expelliarmus",
    "stupefy",
    "inpedimenta",
    "incarcerous"
]
```

```python
jugador1 = input("Escribe tu nombre:")
jugador2 = input("Escribe tu nombre:")
```

```python
while True:

    turno = "jugador1"
    if turno == "jugador1":
        hechizo = input(f" {jugador1} Quieres atacar(A) o defender(D)").lower()
        if hechizo ==  "a":
            hechizo = random.choice(a)
            valores = random.randint(1,15)
            contador2 -= valores
            print(f"Hechizo: {hechizo}, Daño: {valores}, {jugador2}: {contador2} \n {jugador1}:{contador1} \n {jugador2}:{contador2}")
        else:
            hechizo = random.choice(d)
            valores = random.randint(1,15)
            contador1 += valores
            print(f"Hechizo: {hechizo}, Vida: {valores}, {jugador1}: {contador1} \n {jugador1}:{contador1} \n {jugador2}:{contador2}")
        turno = "jugador2"
    elif turno == "jugador2":
        hechizo = input(f" {jugador2} Quieres atacar(A) o defender(D)").lower()
        if hechizo ==  "a":
            hechizo = random.choice(a)
            valores = random.randint(1,15)
            contador1 -= valores
            print(f"Hechizo: {hechizo}, Daño: {valores}, {jugador1}: {contador1} \n {jugador1}:{contador1} \n {jugador2}:{contador2}")
        else:
            hechizo = random.choice(d)
            valores = random.randint(1,15)
            contador2 += valores
            print(f"Hechizo: {hechizo}, Vida: {valores}, {jugador2}: {contador2} \n {jugador1}:{contador1} \n {jugador2}:{contador2}")
        turno = "jugador1"
    if contador1 >= 50:
        print(f"Gana {jugador1}")
        break
    elif contador2 >= 50:
        print(f"Gana {jugador2}")
        break




```
