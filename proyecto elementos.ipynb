opciones de ganar:
el agua gana a fuego, el fuego gana a tierra, tierrra gana a agua



```python
import random    #Importamos el random
```

```python
opciones = ("tierra", "fuego", "agua")     #Las opciones para el random
contador_jugador = 0
contador_python = 0
jugador = input("Ingresa tu nombre:").lower()     #Preguntamos al usuario su nombre

while True:                #Bucle para que se ejecute el juego hasta que el jugador quiera salir
    jugador = input("Elige un elemento:").lower()     
    python = random.choice(opciones)     
    if contador_jugador >= 3 or contador_python >= 3:   #Condicion de parada
        if contador_jugador > contador_python:
            print("¡¡Ganaste!!")
            break
        else:
            print("¡¡Gana Python!!")
            break

    if jugador == python:
        print("empate")
    elif (jugador == "agua" and python == "fuego") or (jugador == "fuego" and python == "tierra") or (jugador == "tierra" and python == "agua"):
        print("gana jugador 1")
    else: 
        print("Gana jugador 2")
```
