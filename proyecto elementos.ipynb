```python
import random    #Importamos el random
opciones = ("tierra", "fuego", "agua")     #Las opciones para el random
opciones_validas = ("tierra", "fuego", "agua", "s")    #Lista con los input's validos
contador_jugador = 0
contador_python = 0             #Contadores para calcular victorias y empates
empate = 0
jugador_nombre = input("Ingresa tu nombre:").lower()     #Preguntamos al usuario su nombre

while True:                #Bucle para que se ejecute el juego hasta que el jugador quiera salir
    jugador = input(f"Elige un elemento: \n Pulsa (S) para salir").lower()      #Eleccion del jugador
    print(f"{jugador_nombre} elige: {jugador}")  
    if jugador not in opciones_validas:                          #Print de control si no selecciona una opcion valida
        print("Opcion no valida.. intentalo de nuevo")
        continue                                    #Para que siga el bucle
    if jugador == "s":                     #Condicion de parada del bucle
        if contador_jugador > contador_python:      #Condicion para la victoria del jugador
            print("¡¡Ganaste!!")
            print(f"Has ganado: {contador_jugador} veces \nPython ha ganado: {contador_python} veces\nHan empatado: {empate} veces")
            print("Hasta la proxima...")
            break
        elif contador_jugador == contador_python:       #Condicion para el empate
            print("¡¡Empate!!")
            print(f"Has ganado: {contador_jugador} veces \nPython ha ganado: {contador_python} veces\nHan empatado: {empate} veces")
            print("Hasta la proxima...")
            break
        else:                                #Condicion para la victoria de python
            print("¡¡Gana Python!!")
            print(f"Has ganado: {contador_jugador} veces \nPython ha ganado: {contador_python} veces\nHan empatado: {empate} veces")
            print("Hasta la proxima... loser")
            break
    python = random.choice(opciones)       #Elige la computadora
    print(f"Python elige: {python}")
        #Condiciones del juego
    if jugador == python:          #Condicion de empate
        print("empate")
        empate += 1
    elif (jugador == "agua" and python == "fuego") or (jugador == "fuego" and python == "tierra") or (jugador == "tierra" and python == "agua"):             #Condicion de victoria para jugador
        contador_jugador += 1
        print(f"!!Gana {jugador_nombre}!!")
    else:                                       #Condicion victoria Python
        contador_python += 1
        print("Gana Python")
```
